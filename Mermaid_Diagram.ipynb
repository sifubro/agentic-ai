```python
from mermaid import Mermaid

diagram = Mermaid("""
sequenceDiagram
  participant Client
  participant AuthManager
  participant DB as SQLite
  participant SessionManager
  participant Orchestrator
  participant Agent
  participant MemoryManager

  Client->>AuthManager: create_user(username,password)
  AuthManager->>DB: INSERT users (user_id)
  Note right of DB: user_id created & stored

  Client->>AuthManager: login(username,password)
  AuthManager->>DB: INSERT tokens (token, refresh_token, user_id, expires_at)
  Note right of DB: token ↔ user_id

  Client->>SessionManager: create_session(token, metadata)
  SessionManager->>AuthManager: validate_token(token)
  AuthManager->>DB: SELECT tokens -> returns user_id
  SessionManager->>DB: INSERT sessions (session_id, user_id, expires_at)

  Client->>Orchestrator: create nodes (node_id)  --(persisted)--> DB
  Orchestrator->>DB: INSERT nodes (node_id, inputs, metadata, status=pending)

  Client->>Orchestrator: execute_graph(session_id)
  Orchestrator->>DB: load nodes (pending)
  Orchestrator->>Agent: Message(from=session_id, task_id=node_id, payload=inputs)
  Agent-->>Orchestrator: response { status, result, memories, next_tasks }
  Orchestrator->>DB: UPDATE nodes (status=result, result)
  Orchestrator->>MemoryManager: add_short(session_id, memories)
  MemoryManager->>DB: INSERT memories (mem_id, user_id, item)
""")

diagram
```
